# 使用web of science进行文献检索

1. 打开 南京大学图书馆网站

2. 打开 电子资源
3. 点击 西文 --> ISI Web of Knowledge平台



4. 点击 進階檢索
5. 输入检索条件：

> TS=(echo chamber) or TS=( group polarization) or TS=(filter bubble) or TS=( tele-cocoon*) or (TS=(information) and TS=(enclave or cocoon*)

6. 点击更多設定--> 選取資料庫--> Web of Science 核心合輯
7. 检索



8. 点击 上图中的数字 40668




9. 点击**新增至勾选的清单**，将全部条目新增至**勾选的清单**！







10. 点击页面右上角勾选的清单，下载文献信息，选取全部条目，但是每次最多下载500条，1-500， 501-1000，依次保存
